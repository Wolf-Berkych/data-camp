### 1. Filter the data for movies released in the 1990s

```python
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
```

### 2. Find the most frequent movie duration

Visualize the data to view the distributions of durations for movies released in the 1990s.

```python
# Calculating the mean, median duration, and the most frequent movie duration

duration_list = movies_90s['duration']

duration_mean = duration_list.mean()
duration_median = duration_list.median()

duration_values = duration_list.value_counts()

most_freq_dur_val = duration_values.idxmax()
most_freq_dur = duration_values.max()

# Building the histogram

plt.hist(duration_list, bins=9, alpha=0.7, color='crimson')

# Signing labels for x-axis, y-axis
plt.title('Durations for movies released in the 1990s.')
plt.xlabel('Duration')
plt.ylabel('Frequency')

# Showing and signing mean and mediad on the plot

plt.axvline(x=duration_mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(x=duration_median, color='b', linestyle='dashed', linewidth=1)

plt.text(duration_mean + 1, 52, 'Mean', color='red', rotation=90)
plt.text(duration_median + 1, 42, 'Median', color='blue', rotation=90)

# As we can notice the most frequent movie duration is 94 minutes for 7 movies
# however th distribution takes the wider range to show frequency
# so we stick to its demonstration (for now)

# Getting the max frequency from bin range

n, bins, patches = plt.hist(duration_list, bins=9, alpha=0.7, color='crimson')
max_frequency = int(max(n))

# Store the max frequency value into duration variable to please Project Instructions

duration = most_freq_dur_val

# Signing and showing it on the plot

plt.axhline(y=max_frequency, color='orange', linestyle='dashed', linewidth=1)
plt.text(2, max_frequency + 0.8 , f'The top frequency: {max_frequency}', color='orange')

plt.show()
```
